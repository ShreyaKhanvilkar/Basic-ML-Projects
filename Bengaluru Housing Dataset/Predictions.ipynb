```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
```

```python
df.total_sqft = df.total_sqft.apply(convert_sqft_to_num)
df = df[df.total_sqft.notnull()]
df.head(2)
```
